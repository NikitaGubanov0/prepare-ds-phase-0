```python
# zip-code представлен в виде чисел
# поэтому мы преобразуем его в число

def fix_zip_code(zip_code):
    try:
        return int(zip_code)
    except Exception as e:
        print(e)
        return np.nan
```

Частая задача, использовать булевые значения для фильтрации данных 

- | `или`,
- & `и`,
- ~ `не`

```python
# только строки с значениями возраста меньше 18 или значения профессии other
df[(df['age'] < 18) | (df['occupation'] == 'other')]  
```

```python
# возвращает только строки с значениями возраста меньше 18 и значения профессии other

df[(df['age'] < 18) & (df['occupation'] == 'other')]  
```

```python
df[~(df['occupation'] == 'other')]  # возвращает только строки с значениями профессии НЕ other
```
